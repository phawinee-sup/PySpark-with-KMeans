```python
# group by: date_hour,station_id
# column: average sensor_value 1-3
df_s123 = df_s12.join(df_s3, ["station_id", 'date'])
df_all = df_s123.groupBy(["date_hour", "station_id"]) \
                .agg(avg("sensor_value1").alias("avg_sensor1"),
                     avg("sensor_value2").alias("avg_sensor2"),
                     avg("sensor_value3").alias("avg_sensor3"))
```

```python
df_all.sort([df_all.date_hour, df_all.station_id]) \
                .toPandas()
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from sklearn.metrics import confusion_matrix
from pyspark.ml.feature import Bucketizer
```

```python
# step 1. specify columns to features
va = VectorAssembler(
        inputCols = ["avg_sensor1", "avg_sensor2", "avg_sensor3"],
        outputCol = "features"
    )

df_train = va.transform(df_all)
df_train.limit(10).toPandas()
```

```python
# step 2. KMeans
km = KMeans().setK(3).setSeed(1)
df_output = km.fit(df_train)
```

```python
# step 3. Show output
df_output.summary.clusterSizes
```

```python
df_output.clusterCenters()
```
